```python
import csv
import copy 
import matplotlib
import pandas as pd

# Allows you to render matplotlib graphs in the same notebook
%matplotlib inline 

def plot_dict(d, label="Please Label Me!"):
    ax = pandas.Series(d).sort_index().plot.bar(color="black", fontsize=16)
    ax.set_ylabel(label, fontsize=16)
```

```python
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data
mappings = process_csv("mapping.csv")
def get_mapping(path):
    mapping_dict = {}
    mappings = process_csv(path)
    for i in mappings:
        mapping_dict[i[0]] = i[1]
    return mapping_dict
mapping = get_mapping("mapping.csv")
def get_raw_movies(path):
    csv_data = process_csv(path)
    header = csv_data[0]
    raw_movies = []
    for i in csv_data[1:]:
        raw_movies_dict = {}
        for key in header:
            value = i[header.index(key)]
            if key == 'title':
                raw_movies_dict.update({key:value})
            if key == 'year':
                raw_movies_dict.update({key:int(value)})
            if key == 'duration':
                raw_movies_dict.update({key:int(value)})
            if key == 'rating':
                raw_movies_dict.update({key:float(value)})
            if key == 'directors' or key == 'actors' or key =='genres':
                value = value.split(', ')
                raw_movies_dict.update({key:value})
        raw_movies.append(raw_movies_dict)
    return raw_movies
def get_movies(movies_path, mapping_path):
    mapps = get_mapping(mapping_path)
    movies = get_raw_movies(movies_path)
    for a in range(len(movies)):
        movie = movies[a]
        movie['title'] = mapps[movie['title']]
        for i in range(len(movie['actors'])):
            movie['actors'][i] = mapps[movie['actors'][i]]
        for x in range(len(movie['directors'])):
            movie['directors'][x] = mapps[movie['directors'][x]]

    return movies
movies = get_movies("movies.csv", "mapping.csv")
movies
```

```python
def search_bar(movies, keyword):
    idx = 0
    while idx < len(movies):
        movie = movies[idx]
        if (keyword not in movie['title']) and (keyword not in movie["genres"]) and  (keyword not in movie["directors"]) and (keyword not in movie["actors"]):
            movies.pop(idx)
        else:
            idx += 1
    return movies
```
